```python
#################################################### Процесс сглаживания #############################################

import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('test.bmp')

# a- Средняя фильтрация Фильтр ни́жних часто́т 
kernel = np.ones((3,3),np.float32)/9
dst = cv.filter2D(img,-1,kernel)
plt.imshow(dst)
plt.xticks([]), plt.yticks([])
plt.show()
cv.imwrite('1.1.jpeg',dst)



kernel = np.array([[1/10,1/10, 1/10], [1/10, 2/10, 1/10], [1/10, 1/10, 1/10]]) 
dst = cv.filter2D(img,-1,kernel)
plt.imshow(dst)
plt.xticks([]), plt.yticks([])
plt.show()
cv.imwrite('1.2.jpeg',dst)
```

```python
#################################################### фильтр нижних частот #############################################

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image


img_src = cv.imread ('test.bmp')


kernel = np.array([[1/16,2/16, 1/16], [2/16, 4/16, 2/16], [1/16, 2/16, 1/16]]) 
kernel = kernel/(np.sum(kernel) if np.sum(kernel)!=0 else 1) 
img_rst = cv.filter2D(img_src,-1,kernel) 
cv.imwrite('2.1.jpeg',img_rst)


kernel = np.array([[2/16,1/16, 2/16], [1/16, 4/16, 1/16], [2/16, 1/16, 2/16]]) 
kernel = kernel/(np.sum(kernel) if np.sum(kernel)!=0 else 1) 
img_rst = cv.filter2D(img_src,-1,kernel) 
cv.imwrite('2.2.jpeg',img_rst)
```

```python
#################################################### фильтр верхних частот #############################################

import numpy as np 
import cv2 
img_src = cv2.imread('test.bmp') 
kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]]) 
kernel = kernel/sum(kernel) 
img_rst = cv2.filter2D(img_src,-1,kernel) 
cv2.imwrite('3.1.jpeg',img_rst)


kernel = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]]) 
kernel = kernel/sum(kernel) 
img_rst = cv2.filter2D(img_src,-1,kernel) 
cv2.imwrite('3.2.jpeg',img_rst)


kernel = np.array([[1, -2, 1], [-2, 5, -2], [1, -2, 1]]) 
kernel = kernel/sum(kernel) 
img_rst = cv2.filter2D(img_src,-1,kernel) 
cv2.imwrite('3.3.jpeg',img_rst)
```
